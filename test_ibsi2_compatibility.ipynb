# Introduction

This notebook tests the agreement of pyCERR's image filter implementations with the [Image Biomarker Standardization Initiative](https://doi.org/10.1148/radiol.231319) (IBSI)'s definitions.

* pyCERR provides 5 convolutional filters for use in radiomics and quantitative image analysis, viz., *mean filters*, *Laplacian-of-Gaussian filters*, *Laws kernels*, *Gabor kernels*, and *wavelet (undecimated, separable) filters*, including 2D/2.5D/3D and rotation-invariant forms, where defined.

* **Phase-1**: pyCERR-computed response maps are compared against consensus reference maps [distributed](https://github.com/theibsi/ibsi_2_reference_data/tree/main/reference_response_maps) by the IBSI using their suggested guideline, i.e., that filter response maps may differ by at most 1% at any voxel from the reference response map.


#### <u>Reference</u>
*Whybra, Philip, et al. "The image biomarker standardization initiative: standardized convolutional filters for reproducible radiomics and enhanced clinical insights." Radiology 310.2 (2024): e231319. doi:10.1148*
